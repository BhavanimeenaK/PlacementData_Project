#Again Data Preprocessing

**Outlier Analysis:**

```python
col=["ssc_p","hsc_p","degree_p","etest_p","mba_p","salary"]
```

```python
for c in col:
  plt.figure()
  sns.boxplot(y=c,data=df)
```

```python
col=["hsc_p","degree_p","salary"]
```

```python
for c in col:
  plt.figure()
  sns.boxplot(y=c, data=df)
  percentile25=df[c].quantile(0.25)
  percentile75=df[c].quantile(0.75)
  iqr=percentile75-percentile25
  upper_limit=percentile75+(1.5*iqr)
  lower_limit=percentile25-(1.5*iqr)
  df=df[(df[c]<=upper_limit) & (df[c]>=lower_limit)]
  plt.figure()
  sns.boxplot(y=c,data=df)
```

*The outliers has been removed.*

**Elbow Method to find the optimal number of clusters:**

```python
inertia=[]
for i in range(1,11):
  print(i)
  kmeans=KMeans(i)
  kmeans.fit(scaled_df)
  print(kmeans.inertia_)
  inertia.append(kmeans.inertia_)
```

```python
Number_of_clusters=range(1,11)
```

```python
plt.plot(Number_of_clusters,inertia,marker="o")
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title("Elbow curve")
```

#Principle Component Analysis

**Dropping the target columns:**

```python
new_df=df.drop(columns=["status","salary"],axis=1)
```

**Standardization:**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit_transform(new_df)
```

**Tranforming data into Principle Components:**

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=5)
pca_array=pca.fit_transform(sc.fit_transform(new_df))
```

**Percentage of data in each Principle Components:**

```python
pca.explained_variance_ratio_
```

*The first PC contains 20%, the second PC contains 14%,the third PC contains 11%,the fourth PC contains 10%,the fifth PC contains 7% of information.*

**Total Percentage of data in all the Principle Components:**

```python
pca.explained_variance_ratio_.sum()
```

*The final PCA contains 64% of information.*

**DataFrame of Principle Components:**

```python
pca_df=pd.DataFrame(pca_array,columns=["PC1","PC2","PC3","PC4","PC5"])
```
